```python
myfile = open('c:/Lab/employees.csv').readlines()
```

```python
emp = myfile[1:]
```

```python
dep_emp = {}

for emp in emp:
    data = emp.split(',')
    dep = data[-1].strip()
    name = data[0]
    try :
        dep_emp[dep].append(name)
    except KeyError:
        dep_emp[dep] = list()
        dep_emp[dep].append(name)

print(dep_emp)
```
